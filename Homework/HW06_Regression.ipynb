Some of you tried to recreate the Extended Data Fig. 2. Here is a variant of (b). I am plotting standard deviation, not confidence interval. Change the below to be confidence interval. Bonus points for telling me why my plot has so many more data points than the one in the paper.

````python
# Determine the mean unique compound count and standard deviation for each group
df_osr = NicholsonKnapp_etal_2023[NicholsonKnapp_etal_2023['crop_type'] == 'OSR']
df_app = NicholsonKnapp_etal_2023[NicholsonKnapp_etal_2023['crop_type'] == 'APP']

mean_app = df_app['uniqueCompound_count'].mean()
mean_osr = df_osr['uniqueCompound_count'].mean()

std_app = df_app['uniqueCompound_count'].std()
std_osr = df_osr['uniqueCompound_count'].std()

# Define numerical labels corresponding to 'APP' and 'OSR'
labels_app = np.ones(len(df_app)) + np.random.normal(0, 0.5, size=len(df_app)) * 0.1
labels_osr = 2 * np.ones(len(df_osr)) + np.random.normal(0, 0.5, size=len(df_osr)) * 0.1

# Set figure size
plt.figure(figsize=(10,6))

# Plot each individual measurement as scatter plot
plt.scatter(labels_app, df_app['uniqueCompound_count'], c='#93be6c', alpha=0.5, edgecolor='black')
plt.scatter(labels_osr, df_osr['uniqueCompound_count'], c='#f6c652', alpha=0.5, edgecolor='black')

# Plot Mean and std dev
plt.errorbar(1, mean_app, yerr=std_app, fmt='o', ecolor='black', elinewidth=2, color='black', mec='black', mfc='#93be6c', ms=15)
plt.errorbar(2, mean_osr, yerr=std_osr, fmt='o', ecolor='black', elinewidth=2, color='black', mec='black', mfc='#f6c652', ms=15)

# label your axes
plt.ylabel('Number of unique compounds')

# Set the x ticks and labels
plt.xlim([0.5,2.5])
plt.xticks([1, 2], ['Apple', 'Oilseed'])

# set a title for your plot
plt.title('Extended Data Fig. 2 b)')

# Show the plot
plt.show()
````
