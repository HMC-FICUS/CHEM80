## Chi-Squared

The chi-squared statistic is like a score that lets us know how far away our observed data is from what we might expect if everything was perfectly uniform or spread out evenly.

Suppose you have a bag of coloured marbles and you expect that the marbles are in equal proportion for each colour. So, if you pick 100 marbles out of the bag, you should, in theory, get an even number of each colour. This is your expected frequency.

Now, when you actually perform this experiment and take out 100 marbles, you count the actual numbers of each colour. This gives you your observed frequency.

The chi-squared statistic is a way to quantify the difference between your observed frequency (how many of each coluor you actually picked) and your expected frequency (how many of each colour you expected to pick if the colors were uniformly distributed).

2. Chi-squared Test: After performing a chi-squared test, you will get a chi-squared statistic and a p-value. The chi-squared statistic measures the difference between the observed frequencies and the frequencies you would expect if the numbers were truly uniformly distributed. A small chi-squared statistic corresponds to observed frequencies close to the expected frequencies, which suggests that the numbers are uniformly distributed.

The p-value gives the probability of seeing the observed (or a more extreme) difference purely by chance, assuming the numbers are truly uniformly distributed. A high p-value (generally, p > 0.05) means that the observed difference is likely due to randomness, suggesting that your numbers are indeed uniformly distributed. Conversely, a low p-value suggests that the numbers are not uniformly distributed.


```python
def uniformity_test(rn):
    # Perform the chi-square test
    hist, bins = np.histogram(rn, bins=10)
    exp_freq = [len(rn) / 10] * 10  # for perfect uniform distribution

    chisq, p = chisquare(hist, f_exp=exp_freq)

    print('Chi-square test statistic:', chisq)
    print('p-value:', p)

uniformity_test(numpy_RNG)
```

With the above in mind, the ideal results for a random number generator would be:
   - Histogram: approximately equal counts in all bins
   - Chi-squared test: low chi-squared statistic
   - p-value: greater than 0.05
